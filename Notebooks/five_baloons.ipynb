```python
import re

# Remove all non-Hebrew characters and replace them with spaces
# Hebrew letters range: \u0590-\u05FF
cleaned_text = re.sub(r'[^\u0590-\u05FF]+', ' ', raw_text)

# Remove extra spaces and strip leading/trailing spaces
cleaned_text = ' '.join(cleaned_text.split())

print(cleaned_text)
```

```python
import random

def monkey_typewriter(target_word):
    """
    Generates random Hebrew letter strings until the target word is matched.
    Returns the number of attempts needed.
    """
    # Hebrew letters range from Unicode 0x05D0 to 0x05EA
    hebrew_letters = [chr(i) for i in range(0x05D0, 0x05EA + 1)]
    
    word_length = len(target_word)
    attempts = 0
    
    while True:
        attempts += 1
        # Generate a random string of the same length
        random_word = ''.join(random.choice(hebrew_letters) for _ in range(word_length))
        
        if random_word == target_word:
            return attempts
```

```python
def computed_monkey_typewriter(target_word):
    """
    Computes the expected number of attempts needed to generate the target word
    by randomly selecting Hebrew letters, without actually generating random strings.
    
    For a word of length n, where each position has 27 possible Hebrew letters,
    the expected number of attempts is 27^n.
    """
    # Hebrew letters range from Unicode 0x05D0 to 0x05EA (27 letters total)
    num_hebrew_letters = 27
    
    word_length = len(target_word)
    
    # Expected number of attempts is num_letters^word_length
    expected_attempts = num_hebrew_letters ** word_length
    
    return expected_attempts
```

```python
# Split cleaned_text into words
words = cleaned_text.split()

words = words[:20]  # Limit to first 20 words for practicality
total_attempts = 0

print("Generating cleaned_text using monkey typewriter...\n")

for i, word in enumerate(words, 1):
    attempts = monkey_typewriter(word)
    total_attempts += attempts
    print(f"Word {i}/{len(words)}: '{word}' generated in {attempts} attempts. Running total: {total_attempts}")

print(f"\nTotal attempts needed: {total_attempts}")
```
