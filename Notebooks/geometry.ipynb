```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure and 3D axes
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the usual xyz axes
axis_len = 1.4
ax.quiver(0, 0, 0, axis_len, 0, 0, arrow_length_ratio=0.08)
ax.quiver(0, 0, 0, 0, axis_len, 0, arrow_length_ratio=0.08)
ax.quiver(0, 0, 0, 0, 0, axis_len, arrow_length_ratio=0.08)

# Label axes
ax.text(axis_len + 0.05, 0, 0, 'x')
ax.text(0, axis_len + 0.05, 0, 'y')
ax.text(0, 0, axis_len + 0.05, 'z')

# Plot the three lines from the origin
points = np.array([
    [1, 1, 0],
    [1, 0.5, 1],
    [-1, 0, 1],
], dtype=float)



colors = ['red', 'green', 'black']
t = np.linspace(-1.2, 1.2, 2)
for p, c in zip(points, colors):
    line = np.outer(t, p)
    ax.plot(line[:, 0], line[:, 1], line[:, 2], linewidth=2, color=c)

# Mark the given points
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=40)

# # Draw the radius-1 sphere in a non-obtrusive way
u = np.linspace(0, 2 * np.pi, 60)
v = np.linspace(0, np.pi, 30)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones_like(u), np.cos(v))
ax.plot_wireframe(x, y, z, linewidth=0.5, alpha=0.18)

# Formatting
ax.set_xlim(-1.3, 1.3)
ax.set_ylim(-1.3, 1.3)
ax.set_zlim(-1.3, 1.3)
ax.set_box_aspect((1, 1, 1))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
