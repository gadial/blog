```python
def coupon_collector(N, cutoff = None):
    if cutoff is None:
        cutoff = N
    return N * sum(1 / (N-k+1) for k in range(1,cutoff+1))
```

```python
for cutoff in range(1,145):
    first_phase = coupon_collector(cutoff)
    remaining = 144 - cutoff
    q, r = remaining // 12, remaining % 12
    second_phase = (12-r)*coupon_collector(q) + r*coupon_collector(q+1)
    result = first_phase + 3*second_phase
    print(cutoff, result)
```

```python
cutoff = 6
first_phase = coupon_collector(cutoff)
print("first phase", first_phase)
remaining = 144 - cutoff
print("remaining", remaining)
q, r = remaining // 12, remaining % 12
print("q,r", q, r)
print("coupon_collector(q)", coupon_collector(q))
print("coupon_collector(q+1)", coupon_collector(q+1))
second_phase = (12-r)*coupon_collector(q) + r*coupon_collector(q+1)
result = first_phase + 3*second_phase
print(cutoff, result)
```
