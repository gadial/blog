```python
np.random.seed(42)
n = 10
p = 0.5
sample_size = 10000
data = [sum([np.random.randint(0, 2) for _ in range(n)]) for _ in range(sample_size)]
counts, bins, patches = plt.hist(data, bins=range(n+2), 
                                 density=True, alpha=0.6, 
                                 color='skyblue', edgecolor='black', 
                                 label='Binomial samples')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y * 1)))
plt.xticks(bin_centers, range(n+1))
#plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: int(y * sample_size)))
# Change y values to probabilities
mu = n * p + 0.5
sigma = np.sqrt(n * p * (1 - p))
x = np.linspace(0, n, 200)  # range for the x-axis
pdf_norm = norm.pdf(x, mu, sigma)

plt.plot(x, pdf_norm, 'r-', lw=2, label='Normal approximation')
plt.show()

print(mu)
print(sigma)
print(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(42)

k = 100
#ticks = range(8)
ticks = range(300,410, 10)
# Plot histogram
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.2%}'.format(y * bin_width)))
plt.xticks(ticks)
x = np.linspace(ticks[0], ticks[-1], 200)  # range for the x-axis

# binomial
p = 0.5
q = 1 - p
n = 7
mu = n*p
sigma = np.sqrt(n*p*q)
mu *= k
sigma *= np.sqrt(k)
pdf_norm = norm.pdf(x, mu, sigma)
plt.plot(x, pdf_norm, 'r-', lw=2)

# dice
sides = 6
results = range(1, sides+1)
p = [1/sides]*sides
mu = sum([result*prob for result, prob in zip(results, p)])
sigma = np.sqrt(sum([prob*(result-mu)**2 for result, prob in zip(results, p)]))
mu *= k
sigma *= np.sqrt(k)
pdf_norm = norm.pdf(x, mu, sigma)
plt.plot(x, pdf_norm, 'b-', lw=2)

# 0-7
a1 = 0
a2 = 7
p = 0.5
q = 1-p
mu = a1*p+a2*q
sigma = np.sqrt(p*(a1-mu)**2+q*(a2-mu)**2)
mu *= k
sigma *= np.sqrt(k)
pdf_norm = norm.pdf(x, mu, sigma)
plt.plot(x, pdf_norm, 'g-', lw=2)

plt.show()
```
