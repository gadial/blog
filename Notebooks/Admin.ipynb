# Create site graph

```python
import os
import re

post_link_regex = re.compile(r'<a href="https?://(?:www\.)?gadial\.net/(?P<year>\d{4})/(?P<month>\d{2})/(?P<day>\d{2})/(?P<name>[^/]+)/">')
item_key_regex = re.compile(r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})-(?P<name>[^/]+)\.md')
item_name_regex = re.compile(r'title: [\'"](.*)[\'"]')
relative_path = os.path.join('..', '_posts')
contents = os.listdir(relative_path)
site_graph = {}
for item in contents:
    item_path = os.path.join(relative_path, item)
    if os.path.isfile(item_path):
        with open(item_path, 'r', encoding='utf-8') as file:
            file_content = file.read()
            key = item_key_regex.match(item).groups()
            name = item_name_regex.search(file_content).group(1)
            links = post_link_regex.findall(file_content)
            site_graph[key] = {'links': links, 'name': name, 'id': len(site_graph)}

```

```python
import json
posts = [{"id": val['id'], "name": val['name']} for val in site_graph.values()]
links = set()
for post in site_graph.values():
    for link_id in post['links']:
        links.add((post['id'], site_graph[link_id]['id']))
links = [{"source":link[0], "target": link[1]} for link in links]

site_map = {"posts": posts, "links": links}
with open("../stuff/nav/site_map.json", 'w', encoding='utf8') as outfile:
    json.dump(site_map, outfile)
```
